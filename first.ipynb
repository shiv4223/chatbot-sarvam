````python
from statsmodels.tsa.arima.model import ARIMA

def forecast_sales(series, steps=1):
    model = ARIMA(series, order=(1,1,1))  
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=steps)
    return forecast

next_month_23rd = forecast_sales(df["23rd"], steps=1)
next_month_24th = forecast_sales(df["24th"], steps=1)

forecast_results = pd.DataFrame({
    "Month": ["Next Month"],
    "Forecasted Sales on 23rd": [next_month_23rd],
    "Forecasted Sales on 24th": [next_month_24th]
})
````

````python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "23rd": [54925, 53096, 61164, 46144, 59685, 57194, 51741, 60457, 43421, 46440, 45819, 40950],
    "24th": [52655, 46973, 46845, 48830, 58922, 51266, 58877, 50956, 41185, 43599, 32590, 33864]
}

df = pd.DataFrame(data)
df["Month_Num"] = range(1, 13)

def create_lag_features(df, column, lag=3):
    for i in range(1, lag+1):
        df[f"{column}_lag_{i}"] = df[column].shift(i)
    return df

df = create_lag_features(df, "23rd")
df = create_lag_features(df, "24th")


df.dropna(inplace=True)

X = df.drop(columns=["Month", "23rd", "24th", "Month_Num"])
y_23rd = df["23rd"]
y_24th = df["24th"]

X_train, X_test, y_train_23rd, y_test_23rd = train_test_split(X, y_23rd, test_size=0.2, random_state=42)
X_train, X_test, y_train_24th, y_test_24th = train_test_split(X, y_24th, test_size=0.2, random_state=42)

model_23rd = XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1)
model_24th = XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1)

model_23rd.fit(X_train, y_train_23rd)
model_24th.fit(X_train, y_train_24th)


latest_data = X.iloc[-1:].values
next_month_23rd = model_23rd.predict(latest_data)[0]
next_month_24th = model_24th.predict(latest_data)[0]



print(next_month_23rd, next_month_24th)
````

````python
import requests

def brave_search(query, api_key, count=10):
    url = "https://api.search.brave.com/res/v1/web/search"
    headers = {"Accept": "application/json", "X-Subscription-Token": api_key}
    params = {"q": query, "count": count}
    
    response = requests.get(url, headers=headers, params=params)
    
    if response.status_code == 200:
        return response.json().get("web", {}).get("results", [])
    else:
        print("Error:", response.status_code, response.text)
        return []

# Replace with your Brave API key
API_KEY = "BSAIiq3fpKRvVEJXIa8W9qRTuIm6ClX"

# Example search query
query = "latest AI trends"
search_results = brave_search(query, API_KEY)

# Display results
for idx, result in enumerate(search_results):
    print(f"{idx+1}. {result['title']}")
    print(result['url'])
    print()
````
